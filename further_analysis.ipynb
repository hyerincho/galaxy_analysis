```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

## Check approximation of $\Omega = \sqrt{8G\rho + \frac{GM_{BH}}{(\Delta x/2)^3}}$ to Keplerian angular velocity

```python
end = 6
def keplerian(x,a,b):
    return a/np.power(x,3./2)+b

r, spVcirc, spVcirc2 = np.load('./run_200610_stdlv09/DD0088_spVcirc.npy')
xdata = np.copy(r[:end])
ydata = np.copy(spVcirc[:end]/(r[:end]))
popt, pcov = curve_fit(keplerian, xdata, ydata)
plt.plot(xdata,ydata,color='g',marker='o',ls='')
rFinest = 0.04/2
plt.plot(rFinest,spVcirc2[0]/(rFinest),marker='x',color='g', label=r'lv09:$\sqrt{8G\rho + \frac{GM_{BH}}{(\Delta x/2)^3}}$')
plt.plot(np.linspace(0,r[end],100), keplerian(np.linspace(0,r[end],100), *popt),'g:')
print(spVcirc2[0]/rFinest/keplerian(rFinest,*popt))

r, spVcirc, spVcirc2 = np.load('./run_200526_lv15local/DD0008_spVcirc.npy')
xdata = np.copy(r[:end])
ydata = np.copy(spVcirc[:end]/(r[:end]))
popt, pcov = curve_fit(keplerian, xdata, ydata)
plt.plot(xdata,ydata,color='b',marker='o',ls='')
rFinest = 0.000625/2
plt.plot(rFinest,spVcirc2[0]/(rFinest),marker='x',color='b', label=r'lv15:$\sqrt{8G\rho + \frac{GM_{BH}}{(\Delta x/2)^3}}$')
plt.plot(np.linspace(0,r[end],100), keplerian(np.linspace(0,r[end],100), *popt),'b:')
print(spVcirc2[0]/rFinest/keplerian(rFinest,*popt))

r, spVcirc, spVcirc2 = np.load('./run_200528_lv16_test/DD0009_spVcirc.npy')
xdata = np.copy(r[:end])
ydata = np.copy(spVcirc[:end]/(r[:end]))
popt, pcov = curve_fit(keplerian, xdata, ydata)
plt.plot(xdata,ydata,color='r',marker='o',ls='')
rFinest = 0.0003125/2
plt.plot(rFinest,spVcirc2[0]/(rFinest),marker='x',color='r', label=r'lv16:$\sqrt{8G\rho + \frac{GM_{BH}}{(\Delta x/2)^3}}$')
plt.plot(np.linspace(0,r[end],100), keplerian(np.linspace(0,r[end],100), *popt),'r:')
print(spVcirc2[0]/rFinest/keplerian(rFinest,*popt))

if 1:
    r, spVcirc, spVcirc2 = np.load('./run_200604_lv17alpha/DD0004_spVcirc.npy')
    xdata = np.copy(r[:end])
    ydata = np.copy(spVcirc[:end]/(r[:end]))
    popt, pcov = curve_fit(keplerian, xdata, ydata)
    plt.plot(xdata,ydata,color='m',marker='o',ls='')
    rFinest = 0.00015625/2
    plt.plot(rFinest,spVcirc2[0]/(rFinest),marker='x',color='m', label=r'lv17:$\sqrt{8G\rho + \frac{GM_{BH}}{(\Delta x/2)^3}}$')
    plt.plot(np.linspace(0,r[end],100), keplerian(np.linspace(0,r[end],100), *popt),'m:')
    print(spVcirc2[0]/rFinest/keplerian(rFinest,*popt))
    
plt.yscale('log')
plt.legend()
plt.xscale('log')
plt.xlabel('radius (kpc)')
plt.ylabel(r'$\Omega$')
```
